```python
import pandas as pd
```

```python
data = pd.read_csv('titanic.csv')
```

```python
data.head()
```

```python
n_data = data.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
```

```python
n_data.head()
```

```python
Ent = n_data.drop('Survived',axis='columns')
Ent.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
Ent['Sex']= le.fit_transform(Ent['Sex'])
```

```python
Ent.isnull().sum()
```

```python
Ent.Age = Ent.Age.fillna(Ent.Age.mean())
```

```python
from sklearn import tree
tr = tree.DecisionTreeClassifier()
```

```python
tr.fit(Ent,n_data.Survived)
```

```python
tr.predict([[3,1,20,71]])==0
```

```python
def predict(cl,gen,age,fare):
    chance = tr.predict([[cl,gen,age,fare]])
    if chance==0:
        print('The Passenger will not Survive.')
    elif chance==1:
        print('The Passenger will Survive.')
    return chance
```

```python
tr.score(Ent,n_data.Survived)
```

```python
for i in range(1,4):
    for j in range(2):
        for k in range(18,26):
            for l in range(50,151):
                print([i,j,k,l])
                predict(i,j,k,l)
```
