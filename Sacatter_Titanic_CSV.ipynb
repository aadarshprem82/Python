```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
```

```python
titanic
```

```python
Age = titanic.Age
Fare = titanic.Fare
```

```python
Age
```

```python
Fare
```

```python
plt.scatter(Age,Fare)
plt.show()
```

```python
plt.figure(figsize=(16,9))
colors = np.random.randint(100, size=(891))
plt.scatter(Age,Fare,c=colors,s=200,alpha=0.5,cmap="nipy_spectral")
plt.xlabel("Age",size=20)
plt.ylabel("Fare",size=20)
plt.show()
```
