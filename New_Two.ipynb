```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a=[3,5,6,7,8,9,4]
b=[3,6,8,9,1,6,7]
plt.plot(a,b)
```

```python
plt.scatter(a,b)
plt.show()
```

```python
plt.scatter(a,b,color='salmon')
plt.show()
```

```python
plt.scatter(a,b,color='salmon',marker='*')
plt.show()
```

```python
plt.scatter(a,b,color='salmon',marker='>')
plt.show()
```

```python
plt.scatter(a,b,color='salmon',marker='<')
plt.show()
```

```python
plt.scatter(a,b,color='salmon',marker='*',s=200)
plt.show()
```

```python
plt.scatter(a,b,color='salmon',marker='H',s=200)
plt.show()
```

```python
plt.scatter(a,b,color='salmon',marker='*',s=200)
plt.scatter(b,a,color='black',alpha=0.3,s=300)
plt.show()
```

```python
slices=[7,2,2,13]
activities = ['sleeping','eating','working','playing']
```

```python
cols=["c","m","r","b"]
```

```python
plt.pie(slices)
plt.show()
```

```python
plt.pie(slices,labels=activities)
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols)
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols,startangle=90)
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols,startangle=60,explode=(0,0,0.2,0))
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols,shadow=True,explode=(0,0,0.2,0),autopct="%1.f%%")
plt.show()
```
